```python
import pandas as pd
```

```python
dowjones=pd.read_csv('indexes/dowjones_data.csv')
nasdaq=pd.read_csv('indexes/nasdaq_data.csv')
nifty=pd.read_csv('indexes/nifty50_data.csv')
sp=pd.read_csv('indexes/sp500_data.csv')
usd=pd.read_csv('indexes/us_dollar_index.csv')
usin=pd.read_csv('indexes/usd_inr_data.csv')

```

```python
d=pd.DataFrame()
```

```python
d['date']=pd.to_datetime(usd['date'])
```

```python
d=pd.merge(d,dowjones,on='date',how='left')
```

```python
d.rename(columns={'close': 'dowjones'}, inplace=True)

    
```

```python
dowjones=dowjones[['date','close']]
```

```python
dowjones['date']=pd.to_datetime(dowjones['date'])
```

```python
nasdaq=nasdaq[['date','close']]
nasdaq['date']=pd.to_datetime(nasdaq['date'])

d=pd.merge(d,nasdaq,on='date',how='left')
d.rename(columns={'close': 'nasdaq'}, inplace=True)

    
```

```python
nifty=nifty[['date','close']]
nifty['date']=pd.to_datetime(nifty['date'])

d=pd.merge(d,nifty,on='date',how='left')
d.rename(columns={'close': 'nifty'}, inplace=True)

    
```

```python
sp=sp[['date','close']]
sp['date']=pd.to_datetime(sp['date'])

d=pd.merge(d,sp,on='date',how='left')
d.rename(columns={'close': 'sp500'}, inplace=True)

    
```

```python
usd=usd[['date','close']]
usd['date']=pd.to_datetime(usd['date'])

d=pd.merge(d,usd,on='date',how='left')
d.rename(columns={'close': 'dollarindex'}, inplace=True)

    
```

```python
usin=usin[['date','close']]
usin['date']=pd.to_datetime(usin['date'])

d=pd.merge(d,usin,on='date',how='left')
d.rename(columns={'close': 'usdinr'}, inplace=True)

    
```

```python
d.to_csv('indexes.csv',index=False)
```
