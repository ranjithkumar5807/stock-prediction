```python
import yfinance as yf
import pandas as pd

# Define tickers
tickers = {
    "dowjones": "^DJI",
    "nasdaq": "^IXIC",
    "nifty": "^NSEI",
    "sp500": "^GSPC",
    "usd_index": "DXY",  # or '^DXY'
    "usd_inr": "USDINR=X"
}

# Download data for each ticker
data = {}
for name, ticker in tickers.items():
    data[name] = yf.download(ticker, start="2000-01-01", end="2024-12-31")

# Access individual dataframes
dowjones = data["dowjones"]
nasdaq = data["nasdaq"]
nifty = data["nifty"]
sp500 = data["sp500"]
usd_index = data["usd_index"]
usd_inr = data["usd_inr"]
```

```python
# Save data to CSV files
for name, df in data.items():
    df.to_csv(f"{name}_dataaaa.csv")
```
